### Fire candidate pixels and extended area

Defining a threshold for radiance and an envelope at a certain distance of the convex hull of all the fire candidate pixels. Ground sampling distance for Hyperion is 30 m, so a deltaR of 10 means 300 m. 

```python
threshRad = 5.
deltaR = 10 
```

Note that the j-index corresponds to the y-coordinate, starting at the top(-left) of the image. 

```python
PRINT = False

i_fire, j_fire = np.where(b220.radiance > threshRad)
firepoints = MultiPoint(zip(j_fire, i_fire))

if PRINT:
    print(zip(i_fire, j_fire))
```

```python
MAKEPLOT = True

hull2 = firepoints.convex_hull
dilated2 = hull2.buffer(deltaR)
patch2 = PolygonPatch(dilated2, facecolor='#6699cc', edgecolor='#6699cc', alpha=0.5, zorder=2)

if MAKEPLOT:
    f, ax = plt.subplots(figsize=(15, 25))
    ax.set_aspect('equal')
    ax.invert_yaxis()
    for p in firepoints:
        ax.plot(p.x, p.y, 'o', color='#999999')
    ax.add_patch(patch2)
```

Calculate the mask of fire candidate pixels from simple thresholding. Then plot the fire mask.

```python
firemask = np.zeros(b220.data.shape)
firemask[b220.radiance > 5.] = 1. 
```

```python
saveout = False
f, ax = plt.subplots(figsize=(15, 25))
#output = ax.pcolor(np.flipud(firemask), cmap='bone', vmin=0, vmax=1)
output = ax.pcolor(firemask, cmap='bone', vmin=0, vmax=1)
patch2 = PolygonPatch(dilated2, facecolor='#6699cc', edgecolor='#6699cc', alpha=0.5, zorder=2)
ax.add_patch(patch2)
ax.set_title("Band 220, Fire Candidates with $L > 5 W / m^2 \mu m$", y=1.08)
ax.set_aspect('equal')
ax.invert_yaxis()
ax.xaxis.set_ticks_position('top')
#ax.set_axis_off()
plt.tight_layout()
if saveout:
    outfile = os.path.join(outdir, "fire01_%s_plot.png" % fireyear)
    plt.savefig(outfile,  bbox_inches='tight', dpi=150)
    plt.close()
```

### Square buffers -- the pedestrian way

Calculate a square buffer around each pixel in the firemask, N pixels in east/west and north/south direction. (E.g., N=10: 300m). Then plot the buffered mask. 

```python
N = 10
firemask_ext = np.zeros(b220.data.shape)
for i, j in zip(i_fire, j_fire):
    firemask_ext[i-N:i+N+1, j-N:j+N+1] = 0.5
firemask_ext[b220.radiance > 5.] = 1. 
```

```python
saveout = False
f, ax = plt.subplots(figsize=(15, 25))
output = ax.pcolor(np.flipud(firemask_ext), cmap='bone', vmin=0, vmax=1)
ax.set_title("Band 220, Fire Candidates and buffer of %s m" % str(N*30))
ax.set_aspect('equal')
#ax.set_axis_off()
plt.tight_layout()
if saveout:
    outfile = os.path.join(outdir, "fire01_buff_%s_plot.png" % fireyear)
    plt.savefig(outfile,  bbox_inches='tight', dpi=150)
    plt.close()
```

### Circular buffers -- the more elegant way

```python
from shapely.geometry import Point
from shapely.ops import cascaded_union
```

A new radius -- a little larger this time. 

```python
deltaCirc = 20
```

```python
firecircles = [Point(j, i).buffer(deltaCirc) for j, i in zip(j_fire, i_fire)]
circbuffer = cascaded_union(firecircles)
```

Plot this one.

```python
saveout = False

firemask = np.zeros(b220.data.shape)
firemask[b220.radiance > 5.] = 1. 

f, ax = plt.subplots(figsize=(15, 25))
#output = ax.pcolor(np.flipud(firemask), cmap='bone', vmin=0, vmax=1)
output = ax.pcolor(firemask, cmap='bone', vmin=0, vmax=1)
ax.set_title("Band 220, Fire Candidates with $L > 5 W / m^2 \mu m$", y=1.08)
ax.set_aspect('equal')
ax.invert_yaxis()
ax.xaxis.set_ticks_position('top')
#ax.set_axis_off()
plt.tight_layout()
for polygon in circbuffer:
    patch = PolygonPatch(polygon, facecolor='#6699cc', edgecolor='#6699cc', alpha=0.5, zorder=2)
    ax.add_patch(patch)
if saveout:
    outfile = os.path.join(outdir, "fire_circ_%s_plot.png" % fireyear)
    plt.savefig(outfile,  bbox_inches='tight', dpi=150)
    plt.close()
```
