The "classical" HFDI isn't usable as band 227 is not calibrated (and doesn't contain usable data). A variant uses band 224, which is calibrated.

```python
#hfdi2 = hypsc.get_normdiff('band224', 'band191')
hfdi2 = hypsc.get_normdiff('band217', 'band196')
```

```python
thresh = -0.1515
np.nansum(hfdi2 > thresh)
```

```python
thresh = -0.1515
b1 = 196
b2 = 217

firemean = -0.052600
fireSDT = 0.048500
thresh_hSTD = firemean - 0.5 * fireSDT

hfdi2 = hypsc.get_normdiff('band217', 'band196')
hfdimask = np.zeros(b220.data.shape)
hfdimask[hfdi2>thresh_hSTD] = 1

saveout = False
f, ax = plt.subplots(figsize=(15, 25))
output = ax.pcolor(np.flipud(hfdimask), cmap='bone', vmin=0, vmax=1)
ax.set_title("HFDI with bands %s and %s" % (b1, b2))
ax.set_aspect('equal')
ax.set_axis_off()
plt.tight_layout()
if saveout:
    outfile = os.path.join(outdir, "HFDI_%s_%s_%s_plot.png" % (b1, b2, fireyear))
    plt.savefig(outfile,  bbox_inches='tight', dpi=150)
    plt.close()
```

## Fire detection with carbon dioxide absorption feature: CIBR

Calculate the carbon-dioxide absoption based fire detection index (continuum interpolated band ratio - CIBR).

```python
cibr = np.divide(hypsc.band185.radiance, 0.666 * hypsc.band183.radiance + 0.334 * hypsc.band189.radiance)
cibr2 = np.divide(hypsc.band185.radiance, 0.666 * hypsc.band183.radiance + 0.334 * hypsc.band188.radiance)
```

```python
print("CIBR min: %.4f, CIBR max: %.4f" % (np.nanmin(cibr), np.nanmax(cibr)))
print("CIBR2 min: %.4f, CIBR2 max: %.4f" % (np.nanmin(cibr2), np.nanmax(cibr2)))
```

Plot CIBR.

```python
thresh_cibr = 0.4
cibr_thresh = np.zeros(b220.data.shape, dtype='uint8')
cibr_thresh[np.logical_and(cibr > thresh_cibr, ~np.isnan(cibr))] = 255 
cibr_thresh_ds = b220.clone(os.path.join(outdir, '%s_cibr_thresh.tif' % firename), cibr_thresh)
```
