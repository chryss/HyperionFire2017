### Make vector files

From now on we work in the native coordinate system of the data (UTM zone 6N). We identify the fire candidate pixels and a convex polygon that covers all of them.

First, section-specific modules are imported.

```python
import fiona
from shapely.geometry import Point, MultiPoint, Polygon, mapping
from shapely.ops import cascaded_union
from descartes.patch import PolygonPatch
```

```python
radthresh = 5.0  # Radiance threshold. In the SWIR, non-fire emitted and reflected radiances are much smaller. 
largebuff = 100 # in m 

i_fire, j_fire = np.where(bd.radiance > radthresh)
e_fire, n_fire = bd.ij2xy(i_fire, j_fire)
# the fire points are in the center of the pixels, while the easting and northing designate the top-left corner
e_fire += 0.5 * bd.delx
n_fire += 0.5 * bd.dely   # note that dely is negative
# Easting and northing as x and y coordinate mesh
E, N = bd._XY
```

### Make raster versions, and save to file

I would also like to have the file in a raster file. This should be a GeoTIFF raster of the same size as the study area file, with two values. Basically, a mask, but marking the inside of the polygons (that is, an inverted mask). I could use 1 for "in the polygon" and 0 for "not in the polygon", but for quick viewing of the file as an 8-bit unsigned integer type GeoTIFF file, it should be easier to multiply the value by 255, so that "in" is displayed as white in default settings. 

To this end, the geometry objects corresponding to the fire points, the circular buffers and the area buffer need to be converted to array coordinates.

Let's define a function that turns an (x, y) coordinate pair into a (j, i) col-row index pair.


```python
def xy2ji(xytup, geotiff):
    restup = geotiff.xy2ij(*xytup)
    return (restup[1], restup[0])
```

We can save and load spectra using pickle. The variable ``specs`` is supposed to be a list or 1D array of spectra, with i and j arrays of the same length of coordinates. 

Some general remarks on persisting the spectra and sample points. There are various options:
* spectra (that is, iterables of spectra) can be saved to file with the pickle module, but the drawback is that he i, j coordinate information is lost; that is acceptable for the fire candidate pixels, which are defined via an external criteron (radiance threshold)
* if samples of coordinate pairs are taken (eg, take a sample of buffer pixels, of outside pixels...) THESE also have to be saved. This can easily be done via json. Then the actual spectra can be pickled. However, saving coordinates and spectra in two different files is a recipe for data corruption
* the overall preferable method would be to save the whole set of arrays (covert any lists to numpay arrays first) to HDF5. ``pygaarst`` contains a function for that purpose in the ``rasterhelpers`` module. It requires numpy arrays of i and j coordinates, and spectra. 

```python
def specstopickle(outfilepath, specs):
    with open(outfilepath, "w") as sink:
        pickle.dump(specs, sink)
```

```python
infilepath = os.path.join(outdir, "%s_firespecs.dat" % firename)

def specsfrompickle(filepath, hypsc, ii, jj, picklefile=True, calculate=False):
    if picklefile:
        try:        
            with open(filepath, "rU") as source:
                specs = pickle.load(source)
        except IOError, EOFError:
            calculate = True
    if (calculate or not picklefile):
        specs = [hypsc.spectrum(i, j) for i, j in itertools.izip(ii, jj)]
        if picklefile:
            specstopickle(filepath, specs)
    return specs
```

## Fit spectra

```python
from sklearn.cluster import MiniBatchKMeans, KMeans
import time
```

```python
batch_size = 20
mbk = MiniBatchKMeans(init='k-means++', n_clusters=6, batch_size=batch_size,
                      n_init=1000, max_no_improvement=1000, verbose=False)
t0 = time.time()
mbk.fit(specs)
t_mini_batch = time.time() - t0
mbk_means_labels = mbk.labels_
mbk_means_cluster_centers = mbk.cluster_centers_
mbk_means_labels_unique = np.unique(mbk_means_labels)
```
