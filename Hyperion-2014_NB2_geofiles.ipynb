# Hyperion fire detection - October 2014

This notebook contains the code to generate the *basic geospatial data files* from the (clipped) Hyperion GeoTIFF scenes. These data files  are then used in the analysis. 

The files in question are:

* Text files containing the list of band names and calibrated wavelengths from Hyperion in general. These are to be used when, for example, importing data into ENVI. They are handy to have. 
* Shapefiles (with georeference, that is, in Earth coordinates, not just in array indices) of the wider study area enclosing all fire candidate pixels.
* Shapefiles of the fire candidate pixels themselves.
* Shapefiles (w/ georeference) of the buffer areas around the fire candidate pixels.
* Visualizations of these files.

Part 1 is required for all other sets of cells, which then can be run independently.

### Make vector files

From now on we work in the native coordinate system of the data (UTM zone 6N). We identify the fire candidate pixels and a convex polygon that covers all of them.

First, section-specific modules are imported.

```python
import fiona
from shapely.geometry import Point, MultiPoint, mapping
from shapely.ops import cascaded_union
from descartes.patch import PolygonPatch
```

```python
radthresh = 5.0  # Radiance threshold. In the SWIR, non-fire emitted and reflected radiances are much smaller. 
largebuff = 100 # in m 

i_fire, j_fire = np.where(bd.radiance > radthresh)
e_fire, n_fire = bd.ij2xy(i_fire, j_fire)
# the fire points are in the center of the pixels, while the easting and northing designate the top-left corner
e_fire += 0.5 * bd.delx
n_fire += 0.5 * bd.dely   # note that dely is negative
# Easting and northing as x and y coordinate mesh
E, N = bd._XY
```

```python
# calculate fire shapes
firepoints = MultiPoint(zip(e_fire, n_fire))
hull = firepoints.convex_hull
dilated = hull.buffer(largebuff)
```

```python
MAKEPLOT = True
if MAKEPLOT:
    f, ax = plt.subplots(figsize=(15, 25))
    output = ax.pcolor(E, N, np.flipud(bd.radiance), cmap='hot', vmin=0, vmax=70)
    ax.set_aspect('equal')
    plt.tight_layout()
    patch2 = PolygonPatch(dilated, facecolor='#6699cc', edgecolor='#6699cc', alpha=0.4, zorder=2)
    ax.add_patch(patch2)
    for p in firepoints:
        ax.plot(p.x, p.y, 'o', color='#999999', alpha=0.7)
```

Now let's make circular buffers around the points and define their union as the close-proximity vector collcetion. The result is a multipolygon object. 

```python
buffrad = 600 # in m
firecircles = [Point(eastx, northy).buffer(buffrad) for eastx, northy in zip(e_fire, n_fire)]
circbuffer = cascaded_union(firecircles)
```

```python
circbuffer.geom_type
```

Let's plot this one.

```python
saveout = False

f, ax = plt.subplots(figsize=(15, 25))
output = ax.pcolor(E, N, np.flipud(bd.radiance), cmap='hot', vmin=0, vmax=70)
ax.set_aspect('equal')
plt.tight_layout()
#ax.set_title("Band 220, Fire Candidates with $L > 5 W / m^2 \mu m$", y=1.08)
for polygon in circbuffer:
    patch = PolygonPatch(polygon, facecolor='#6699cc', edgecolor='#6699cc', alpha=0.4, zorder=2)
    ax.add_patch(patch)
if saveout:
    outfile = os.path.join(outdir, "fire_circ_%s_plot.png" % fireyear)
    plt.savefig(outfile,  bbox_inches='tight', dpi=150)
    plt.close()
```

### Save vectors to file

We need a coordinate reference system for the shapefile. We get it from the raster data object. The `fiona` module contains a function to translate it from the PROJ4 string to the mapping format that is needed.

```python
from fiona.crs import from_string
my_crs = from_string(bd.proj4)
```

Next we need a schema. We use the simplest one. 

Fire pixels first.

```python
schema = { 'geometry': 'Point', 'properties': {'id': 'int'} }
```

```python
outfn = os.path.join(basedir, filedir, outgis, 'firepixels_{0}.shp'.format(firename))

with fiona.open(
    outfn, "w", driver="ESRI Shapefile", crs=my_crs, schema=schema) as output:
    for idx, item in enumerate(firepoints):
        output.write({
            'geometry': mapping(item),
            'properties': {'id': idx}
        })
```

Polygons next.

```python
schema = { 'geometry': 'Polygon', 'properties': {'id': 'int'} }
```

```python
outfn = os.path.join(basedir, filedir, outgis, 'convexfirehull_{0}.shp'.format(firename))

with fiona.open(
    outfn, "w", driver="ESRI Shapefile", crs=my_crs, schema=schema) as output:
    output.write({
        'geometry': mapping(dilated),
        'properties': {'id': 0}
    })
```

```python
outfn = os.path.join(basedir, filedir, outgis, 'circbuffer_{0}.shp'.format(firename))

with fiona.open(
    outfn, "w", driver="ESRI Shapefile", crs=my_crs, schema=schema) as output:
    for idx, item in enumerate(circbuffer):
        output.write({
            'geometry': mapping(item),
            'properties': {'id': idx}
        })
```
